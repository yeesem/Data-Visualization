```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df3 = pd.read_csv('https://raw.githubusercontent.com/yeesem2002/Machine-Learning-Datasets/main/df3')
```

```python
df3.head()
```

```python
df3.info()
```

```python
df3.plot.scatter(x='a',y='b',c='green',figsize = (12,3),s=50)
```

```python
df3['a'].hist(bins=20,color='blue')
```

```python
plt.style.use('ggplot')
df3['a'].plot.hist(bins=30,alpha=0.5)
```

```python
df3[['a','b']].plot.box()
```

```python
df3['d'].plot.kde()
```

```python
df3['d'].plot.kde(lw=3,ls='--')
```

```python
df3.loc[0:30].plot.area(alpha=0.5)
```

```python
f = plt.figure()
df3.loc[0:30].plot.area(alpha=0.5)
plt.legend(loc='center left',bbox_to_anchor=(1.0,0.5))
plt.show()
```

```python
df3.loc[0:30].plot.line()
```
