```r
flights %>%
     drop_na() %>%
      arrange(arr_delay) %>%
         select(tailnum) %>%
                       head(1)
```

```r
flights %>%
  mutate(flight_time_min = hour*60 + minute)
```

```r
new_flights = flights %>% select(dep_delay,arr_delay,distance,air_time)
```

```r
new_flights = new_flights %>% mutate(gain_time = dep_delay - arr_delay)
```

```r
new_flights = new_flights %>% mutate(speed = distance / air_time * 60)
head(new_flights,1)
```

```r
new_flights2 = flights %>% select(dep_delay,arr_delay,distance,air_time)
new_flights2 = new_flights %>% mutate(gain_time = dep_delay - arr_delay,speed = distance / air_time * 60)
head(new_flights2)
```

```r
#Need to remove na first before calculate mean
#n() - calculate numnber of rows
flights %>%
  summarise(mean_air_time = mean(air_time,na.rm=TRUE),n=n())
```

```r
#summarise() and group_by()

flights %>%
filter (origin == "JFK" & dest == "LAX") %>%
group_by (tailnum) %>%
summarise (mean_air_time = mean(air_time, na.rm = TRUE), n=n()) %>%
head()
```

```r
#use group_by() and ungroup() without summarise()

flights %>%
drop_na() %>%
group_by(dest, origin) %>%
mutate(
    air_time_mean = mean(air_time),
    air_time_sd = sd(air_time),
    n = n()
  ) %>%
  ungroup() %>%
filter(origin=='EWR' & dest=='IAH') %>%
head()
```

```r
flights %>%
  group_by(dest) %>%
  summarise(
    count = n(),
    dist = mean(distance, na.rm = TRUE),
    delay = mean(arr_delay, na.rm = TRUE)
 )
```

```r
flights %>%
  mutate(long_flight = air_time >= 360)  %>%
    count(long_flight)
```

```r
flights %>%
  group_by(tailnum) %>%
    mutate(long_flight = air_time > 360)  %>%
      count(long_flight = TRUE)  %>%
       arrange(desc(n))
```

```r
top_carrier <- flights %>%
  group_by(carrier) %>%
     filter(n() >= 10000) %>%
       ungroup()

head(top_carrier)
```

```r
#Create a new column basis count option
flights %>%
  mutate(long_flight = (air_time >= 6 * 60)) %>%
  View()
```
