```python
import seaborn as sns
%matplotlib inline
```

# **Distribution plots**

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#kde indicate the line
sns.distplot(tips['total_bill'],kde = False)
```

```python
sns.distplot(tips['total_bill'],kde = True)
```

```python
sns.displot(tips['total_bill'],kde=False,bins=30)
```

```python
#x-var,y-var,dataset
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
#kind = 'hex' - present the data in the size of hexagon
sns.jointplot(x='total_bill',y='tip',data = tips,kind = 'hex')
```

```python
#reg indicates the regression
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y = 'tip',data = tips,kind = 'kde')
```

```python
#Relationship between all the variables
sns.pairplot(tips)
```

```python
#Hue colours the sex to differentiate between the male and female
sns.pairplot(tips,hue='sex')
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.kdeplot(tips['total_bill'])
```

# **Matrix Plots**

```python
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
tips.head()
```

```python
flights.head()
```

```python
tips.corr()
```

```python
tc = tips.corr()
sns.heatmap(tc)
```

```python
tc
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
flights
```

```python
fp = flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fp.corr())
```

```python
sns.heatmap(fp,cmap='magma',annot = False,linewidths=3,linecolor='white')
```

```python
#Cluster map
sns.heatmap(fp)
```

```python
sns.clustermap(fp,cmap = 'coolwarm')
```

# **Grids**

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris,hue='species')
```

```python
#Custormize the pairplot by the users
sns.PairGrid(iris)
```

```python
import matplotlib as plt
g = sns.PairGrid(iris)
g.map_diag(sns.histplot)
g.map_upper(sns.scatterplot)
g.map_lower(sns.kdeplot)
```

# **Regression Plots**
