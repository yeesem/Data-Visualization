```python
import seaborn as sns
%matplotlib inline
```

# **Distribution plots**

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
#kde indicate the line
sns.distplot(tips['total_bill'],kde = False)
```

```python
sns.distplot(tips['total_bill'],kde = True)
```

```python
sns.displot(tips['total_bill'],kde=False,bins=30)
```

```python
#x-var,y-var,dataset
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
#kind = 'hex' - present the data in the size of hexagon
sns.jointplot(x='total_bill',y='tip',data = tips,kind = 'hex')
```

```python
#reg indicates the regression
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y = 'tip',data = tips,kind = 'kde')
```

```python
#Relationship between all the variables
sns.pairplot(tips)
```

```python
#Hue colours the sex to differentiate between the male and female
sns.pairplot(tips,hue='sex')
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.kdeplot(tips['total_bill'])
```

# **Categorical Plot**

```python
sns.barplot(data=tips)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
import numpy as np
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex')
```

```python
#Combination of violinplot and stripplot
#Dont use this for large dataset
#Dont scale very well for large dataset
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```
